## 1. Basic Stock Price Analysis (SOFI)

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Pick a company
ticker = "SOFI"

# Get data
data = yf.download(ticker, start="2023-01-01", end="2025-01-01")

# Show first rows
print(data.head())

# Plot closing price
data["Close"].plot(title=f"{ticker} Stock Price")
plt.show()
```

## 2. Comparing SOFI vs S&P 500

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

tickers = ["SOFI", "^GSPC"]
data = yf.download(tickers, start="2023-01-01", end="2025-01-01")

normalized = data["Close"] / data["Close"].iloc[0] * 100

normalized.plot(figsize=(10,6))
plt.title("SOFI vs S&P 500 (Normalized Performance)")
plt.ylabel("Normalized Price (Start = 100)")
plt.show()
```

This chart compares SOFI to the S&P 500 by setting both to the same starting value.

From this, I can start to see whether SOFI is outperforming or underperforming the broader market.

My first impression is that SOFI is more volatile than the market, which could mean it has a stronger reaction to specific news regarding the company and investor sentiment.

## 3. Volatility Comparison

```python
import numpy as np

returns = data["Close"].pct_change()
volatility = returns.std()

print("Volatility (Daily Standard Deviation):")
print(volatility)
```

Volatility measures how much a stock moves from day to day.

A higher volatility number means the stock has larger price swings and may be riskier.

If SOFI has a higher volatility than the S&P 500, that supports the idea that SOFI is more sensitive to company-specific news and investor sentiment than the broader market.

## 4. Returns Analysis

Here I want to understand how much SOFI has actually returned over time compared to the market.

```python
# Cumulative returns
cumulative_returns = (1 + returns).cumprod()

cumulative_returns.plot(title="Cumulative Returns: SOFI vs S&P 500")
```

Cumulative returns show how an initial investment would have grown over time.

This helps compare the actual performance instead of just price levels.

If SOFI outperforms the S&P 500, it could offer higher return potential.
If it underperforms, it suggests there's weaker long-term performance or higher volatility without reward.

## 5. Risk vs Return Analysis

In this section, I compare SOFI’s returns to its volatility to understand whether the higher risk is justified by higher returns.

```python
import pandas as pd
import yfinance as yf

# Define tickers
tickers = ["SOFI", "^GSPC"]

# Download data for each ticker separately
sofi_data = yf.download("SOFI", start="2023-01-01", end="2025-01-01")
gspc_data = yf.download("^GSPC", start="2023-01-01", end="2025-01-01")

# Extract 'Close' prices and assign names to the Series
sofi_close_prices = sofi_data["Close"]
sofi_close_prices.name = "SOFI"
gspc_close_prices = gspc_data["Close"]
gspc_close_prices.name = "^GSPC"

# Combine into a single DataFrame
data = pd.concat([sofi_close_prices, gspc_close_prices], axis=1).dropna()

# Calculate daily returns
returns = data.pct_change()

# Mean daily return
mean_returns = returns.mean()

# Risk (volatility)
risk = returns.std()

# Combine into one table
summary = pd.DataFrame({
    "Mean Return": mean_returns,
    "Volatility": risk
})

print(summary)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a scatter plot for Risk vs Return
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Volatility', y='Mean Return', data=summary, hue=summary.index, s=200)

# Annotate each point with its ticker symbol
for i, txt in enumerate(summary.index):
    plt.annotate(txt, (summary['Volatility'].iloc[i], summary['Mean Return'].iloc[i]), xytext=(5, 5), textcoords='offset points')

plt.title('Risk vs. Return: SOFI vs. S&P 500')
plt.xlabel('Volatility (Standard Deviation of Daily Returns)')
plt.ylabel('Mean Daily Return')
plt.grid(True)
plt.axhline(0, color='grey', linestyle='--') # Add a horizontal line at 0 return
plt.axvline(0, color='grey', linestyle='--') # Add a vertical line at 0 volatility (though not typically reached)
plt.show()
```

This table (with associated visual aid) compares return and risk for SOFI and the S&P 500.

If SOFI has:
- higher volatility but not significantly higher returns → it may not be worth the extra risk
- higher returns AND higher volatility → it could be a higher-risk, higher-reward investment
- lower returns and higher volatility → it is inefficient and unattractive compared to the market

Based on my results:
For the period analyzed, SOFI demonstrates a higher mean daily return (0.003188) but also significantly higher volatility (0.038448) compared to the S&P 500 (mean return: 0.000892, volatility: 0.008106). This suggests that SOFI is a higher-risk, higher-reward investment. Investors in SOFI would have experienced larger price swings, but also a greater average daily return over this specific timeframe.
